# Data preparation for auto-labelling

Define dataset path

```python
# Define dataset path
dataset_path = r"/MS3D/data/2024_04_03"
```

Make train.txt and val.txt

```python
import os, sys
# Make ImageSets dir and append
if "ImageSets" not in os.listdir(dataset_path):
    os.mkdir(dataset_path + "/ImageSets")
sys.path.append(os.path.join(dataset_path, "ImageSets"))

# Read Sequences
sequences = os.listdir(dataset_path + "/sequences")

# Write to train.text
with open(dataset_path + '/ImageSets/train.txt', "w") as file:
    for sequence in sequences:
        file.write(sequence + "\n")

# Write empty val.txt
with open(dataset_path + "/ImageSets/val.txt", "w") as file:
    file.write("")
```

Copy generate lidar odom file and generate odometry

```python
# Copy gen_odometry_file to correct folder:
import shutil
shutil.copyfile("/MS3D/tools/generate_lidar_odom.sh", dataset_path + "/generate_lidar_odom.sh")

cmd = f"cd {dataset_path} && ./generate_lidar_odom.sh"
!{cmd}
```

Generate infos .pkl files

```python
cmd = f"cd /MS3D && \
        python3 -m pcdet.datasets.custom.custom_dataset create_infos /MS3D/data/2024_04_03"
!{cmd}
```
